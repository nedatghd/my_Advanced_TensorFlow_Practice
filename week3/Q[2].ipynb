- lambda_function

```python
scale = tf.Variable(0.01)
threshold = tf.Variable(0.0)
Leaky_Relu = tf.keras.layers.Lambda(lambda x: tf.where(x>threshold, x, scale*x))
```

```python
#Sanity-cheak
output = Leaky_Relu([-3.0, -1.0, 0.0, 2.0])
list(output.numpy())
```

- OOP

```python
class LeakyRelu(tf.keras.layers.Layer):
  def __init__(self, scale, threshold):
    super(LeakyRelu, self).__init__()
    self.scale = tf.Variable(0.01)
    self.threshold = tf.Variable(0.0)

  def call(self, inputs):
    return tf.where(inputs > self.threshold, inputs, self.scale*inputs)
```

```python
#Sanity Cheak
input = tf.linspace(start=-5, stop=5, num=1000)
output = LeakyRelu(0.01,0.0)([input])
output = output.numpy()
```

```python
plt.plot(input, output[0])
plt.title('Leaky_Relu activation function')
```

---



- Lambda function:

```python
swish_function = tf.keras.layers.Lambda(lambda x :x * tf.math.pow(1+tf.math.exp(-x),-1))
```

```python
#Sanity-cheak
output = swish_function(input)
output = output.numpy()
#print(output,type(output))
plt.plot(input, output)
plt.title('Gelu activation Function')
```

- OOP

```python
class SwishActivation(tf.keras.layers.Layer):
  def __init__(self):
    super(SwishActivation, self).__init__()
  def call(self,inputs):
    return inputs * tf.math.pow(1+tf.math.exp(-inputs),-1)
```

```python
SwishActivation()(input)
```

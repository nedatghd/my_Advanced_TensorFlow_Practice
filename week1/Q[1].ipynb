```python
class IntegerCalculator():
  def __init__(self,num1,num2):
    self.num1 = num1
    self.num2 = num2
  # adding
  def add(self):
    return self.num1 + self.num2 
  # subtracting
  def subtract(self):
    return self.num1 - self.num2 
  # multiplication
  def multiply(self):
    return self.num1 * self.num2 
  # division
  def divide(self):
    return self.num1 / self.num2 

# Sanity Cheack
if __name__ == '__main__':
  my_claculator = IntegerCalculator(2,5)
  print(my_claculator.add())
  print(my_claculator.subtract())
  print(my_claculator.multiply())
  print(my_claculator.divide())
```

```python
class Student():
  def __init__(self ,name ,num_hw ,course):
    self.name = name
    self.num_hw = num_hw
    self.course = course
  def info(self):
    print('{} is {} Student. {} HWs Received.'.format(self.name ,self.course ,self.num_hw))
  def score(self): 
    print('{}s score is: {}.'.format(self.name ,int(self.num_hw*100/6)))
  def certificate(self):
    if self.num_hw > 4 :
      print('{} is qualified to get {} certificate with score {}/100.'.format(self.name ,self.course ,int(self.num_hw*100/6)))
    else:
      print('{} is not qualified to get {} certificate.More HW is needed.'.format(self.name ,self.course))


# Sanity Cheack
if __name__ == '__main__':

  person1 = Student('Ali' ,5 ,'Advanced TF')
  person2 = Student('Reza' ,2 ,'Advanced TF')

  person1.info()
  person1.score()
  person1.certificate()
  print('**************************')
  person2.info()
  person2.score()
  person2.certificate()  
```

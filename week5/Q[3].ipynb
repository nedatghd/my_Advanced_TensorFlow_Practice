```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
def grad_tape(x,func):
    with tf.GradientTape(persistent=True) as tape:
        tape.watch(x)
        y=func(x)
    dy_dx=tape.gradient(y,x)
    func_name=f'{func}'.split(' ')[1]
    plt.plot(x,y,label=f'{func_name}')
    plt.plot(x,dy_dx,label=f'derivitive of {func_name}')
    plt.legend()
```

```python
x=tf.convert_to_tensor(np.arange(-10,10,0.01))
```

```python
grad_tape(x,tf.nn.relu)
```

```python
grad_tape(x,tf.math.tanh)
```

```python
grad_tape(x,tf.math.sigmoid)
```
